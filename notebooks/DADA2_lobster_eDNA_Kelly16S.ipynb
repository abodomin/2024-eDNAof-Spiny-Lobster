```python
#creating phylogenetic tree
#align sequences
!qiime alignment mafft \
--i-sequences /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/seq_ASVs_lobC16S_euks_fwd.qza \
--o-alignment /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/aligned_seq_ASVs_lobC16S_euks_fwd.qza

#removing hypervariable or uninformative positions 
#that might interfere with accurate phylogenetic inference
!qiime alignment mask \
--i-alignment /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/aligned_seq_ASVs_lobC16S_euks_fwd.qza \
--o-masked-alignment /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/masked_aligned_seq_ASVs_lobC16S_euks_fwd.qza

#Builds an unrooted phylogenetic tree using the FastTree algorithm
!qiime phylogeny fasttree \
--i-alignment /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/masked_aligned_seq_ASVs_lobC16S_euks_fwd.qza \
--o-tree /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/unrooted_masked_aligned_seq_ASVs_lobC16S_euks_fwd.qza

#Roots the unrooted tree using midpoint rooting — 
#places the root at the midpoint of the longest distance between two tips in the tree.
!qiime phylogeny midpoint-root \
--i-tree /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/unrooted_masked_aligned_seq_ASVs_lobC16S_euks_fwd.qza \
--o-rooted-tree /Users/aiko.abo.dominguez/Desktop/PAV1_eDNA_project/QIIME2_code/rooted_masked_aligned_seq_ASVs_lobC16S_euks_fwd.qza
```
